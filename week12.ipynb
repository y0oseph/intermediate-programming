# Branching

When we want to develop the code, we may want to leave the main code safe and functional until we are finished developping. To do this we "branch" away from the main code, work within this new branch, and then rejoin the main master branch when we are finished (through something called a "pull request" or "PR")

# Group Coding

If you have multiple poeple who want to work on the code together there are two options:

* Collaborating: Add multiple people to the same repo.

* Forking: Multiple people have copied repos, which they can then merge back to yours later.

Collaborating is good for smaller projects with people you trust.

Forking is good for bigger projects in which there may be collaborations from strangers.

# Reverting

Things can get messy. If you ever want to revert the master branch to a previous commit:

1) Create a new branch (ie. "new_branch")
2) In the "Source Control" menu you have a graph with lots of nodes representing each commit. When you hover over one of them you get a little pop-up showing info about the commit. In the bottom left of this you see a code representing the commit (actually called a hash). Copy this.
3) Then open the terminal and type:
"git revert --hard " followed by the copied hash.

Then:

git merge --strategy=ours master

git checkout master

git merge new_branch

### Nuclear option:

If things are really messed up, you can delete the repo from your local machine and clone it again from the source.

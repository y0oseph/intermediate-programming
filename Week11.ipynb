### Exercise 1

Develop a function "printEmployeeInfo" that takes an employee name and a list containing two dictionaries.

One dictionary holds the names of the employees (keys) and their respective roles in the company (values).

The other dictionary holds the company roles (keys) and their respective salaries (values).

The function should print out both the role and the salary of the employee.


```python
def printEmployeeInfo(dict_list, employee):
    # Add your code here (and see test case in the next cell)
    jobs = dict_list[0]
    salaries = dict_list[1]

    role = jobs[employee]
    salary = salaries[role]


    print(employee + " works in " + role + " and has a salary of " + str(salary))
    return None
```

```python
jobs = {"anne":"IT", "sue":"software", "jim":"HR", "evan":"teacher"}

salaries = {"IT":100000, "software":200000, "HR":80000, "teacher":5}

dict_list = [jobs, salaries]

printEmployeeInfo(dict_list, "jim")
```

### Some notes about updating dictionaries in a loop

```python
# For normal lists, it's usually not a problem to change the list while you loop

a = [1,2,3,4,5]

for n in a:
    if n == 4:
        a.append(6)  # Here we extend the list while in a loop!

a  # But it works just fine
```

```python
# But for dictionaries this breaks things:

dict = {"England":"London", "France":"Paris", "Germany": "Munich"}

for key in dict:
    if dict[key][0] == "L":
        dict["Spain"] = "Madrid"

dict
```

```python
# We can fix this by using "dict.copy()"

for key in dict.copy():
    if dict[key][0] == "L":
        dict["Spain"] = "Madrid"

dict
```

### Exercise 2

Make a function that takes in a sentence and a word, and returns how many times that word is found in the sentence.

Hint: Remove punctuation, make everything lower case, then split the sting into a list where each entry is a word.

```python
# String splitting

def howManyInstances(string, word):
    # Add your code here
    count = 0
    x = string.split(" ")
    
    for words in x:
        if words.lower() == word:
            count += 1

    return count

print(howManyInstances("The fox had fun with the dog and the cat.", "the"))
```
